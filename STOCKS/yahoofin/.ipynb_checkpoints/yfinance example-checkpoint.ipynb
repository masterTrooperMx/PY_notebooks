```python
tickers = ['AAPL', 'BTC', 'ALFAA.MX', 'LLY.MX']
for ticker in tickers:
    dat = yf.Ticker(ticker)
    try:
        income_stmt = dat.quarterly_income_stmt
        ebitda = income_stmt.loc['EBITDA']  # Esto devuelve una Serie por trimestre
        latest_ebitda = ebitda.iloc[0] if not ebitda.empty else None
        print(f"{ticker} - √öltimo EBITDA reportado: {latest_ebitda}")
    except KeyError:
        print(f"{ticker} - EBITDA no disponible")
    except Exception as e:
        print(f"{ticker} - Error: {e}")
```

```python
#import yfinance as yf
#import pandas as pd

tickers = ['AAPL', 'PFE.MX', 'ALFAA.MX', 'LLY.MX', 'PDD', 'MSTR', 'MCD', 'BAI', 'TSM', 'GENTERA.MX', 'BIMBO.MX']

for ticker in tickers:
    dat = yf.Ticker(ticker)
    try:
        info = dat.quarterly_income_stmt
        iprice = dat.analyst_price_targets

        # precio accion
        price = iprice["current"]
        # Extrae los valores de forma segura
        revenue = info.loc["Total Revenue"].dropna().iloc[0] if "Total Revenue" in info.index else None
        ebitda = info.loc["EBITDA"].dropna().iloc[0] if "EBITDA" in info.index else None
        net_income = info.loc["Net Income"].dropna().iloc[0] if "Net Income" in info.index else None
        eps = info.loc["Diluted EPS"].dropna().iloc[0] if "Diluted EPS" in info.index else None

        # Verifica si todos los valores est√°n disponibles
        if None in (revenue, ebitda, net_income, eps):
            print(f"‚ö†Ô∏è {ticker}: Datos insuficientes para c√°lculo")
            continue

        margen_ebitda = ebitda / revenue
        margen_neto = net_income / revenue
        pu = dat.history(period="1d")["Close"][-1] / eps

        print(f"\nüìä Acci√≥n: {ticker}")
        print(f"‚Ä¢ Precio: ${price:.2f}")
        print(f"‚Ä¢ Margen EBITDA: {margen_ebitda:.2%}")
        print(f"‚Ä¢ Margen Neto: {margen_neto:.2%}")
        print(f"‚Ä¢ UPA o BPA prom: ${eps:.2f}")
        print(f"‚Ä¢ P/U: {pu:.2f} years")
    except Exception as e:
        print(f"\n‚ùå {ticker} - Error: {e}")
```

```python
tickers_mx = ['IVVPESOISH.MX', 'FMX23.MX', 'LABB.MX', 'CUERVO.MX', 'GRUMAB.MX', 'VESTA.MX', 'GENTERA.MX', 'NUTRISAA.MX']
tickers_us = ['BTC', 'RIOT', 'ORCL', 'GOOX', 'NFLX', 'NVDA', 'AMD', 'TGT', 'BRK.B', 'AVGO', 'DIA', 'TSLA', 'XXII', 'ETH']

def imprime_ticker_info(tickers):
    for ticker in tickers:
        dat = yf.Ticker(ticker)
        try:
            info = dat.quarterly_income_stmt
            iprice = dat.analyst_price_targets

            # precio accion
            price = iprice["current"]
            # Extrae los valores de forma segura
            revenue = info.loc["Total Revenue"].dropna().iloc[0] if "Total Revenue" in info.index else None
            ebitda = info.loc["EBITDA"].dropna().iloc[0] if "EBITDA" in info.index else None
            net_income = info.loc["Net Income"].dropna().iloc[0] if "Net Income" in info.index else None
            eps = info.loc["Diluted EPS"].dropna().iloc[0] if "Diluted EPS" in info.index else None

            # Verifica si todos los valores est√°n disponibles
            if None in (revenue, ebitda, net_income, eps):
                print(f"‚ö†Ô∏è {ticker}: Datos insuficientes para c√°lculo")
                continue

            margen_ebitda = ebitda / revenue
            margen_neto = net_income / revenue
            pu = dat.history(period="1d")["Close"][-1] / eps

            print(f"\nüìä Acci√≥n: {ticker}")
            print(f"‚Ä¢ Precio: ${price:.2f}")
            print(f"‚Ä¢ Margen EBITDA: {margen_ebitda:.2%}")
            print(f"‚Ä¢ Margen Neto: {margen_neto:.2%}")
            print(f"‚Ä¢ UPA o BPA prom: ${eps:.2f}")
            print(f"‚Ä¢ P/U: {pu:.2f} years")
        except Exception as e:
            print(f"\n‚ùå {ticker} - Error: {e}")

imprime_ticker_info(tickers_mx)
imprime_ticker_info(tickers_us)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Datos combinados (resumen de tus portafolios)
data = {
    'Ticker': ['LABB.MX', 'CUERVO.MX', 'GRUMAB.MX', 'VESTA.MX', 'NUTRISAA.MX',
               'RIOT', 'ORCL', 'NFLX', 'NVDA', 'AMD', 'TGT', 'AVGO', 'TSLA', 'XXII'],
    'P_U': [48.75, 38.13, 818.68, 1750.15, -46.00,
            48.88, 196.79, 17.77, 137.60, 271.71, 58.03, 400.24, 1002.79, -2.96],
    'Margen_Neto': [7.59, 17.48, 8.38, 41.20, -4.75,
                    57.97, 19.61, 22.13, 55.98, 13.44, 3.71, 25.95, 4.89, 366.42],
    'Margen_EBITDA': [18.03, 29.05, 18.14, 98.85, -6.04,
                      109.39, 41.01, 64.07, 67.97, 22.78, 8.35, 51.99, 13.03, -195.19],
    'Mercado': ['BMV', 'BMV', 'BMV', 'BMV', 'BMV',
                'USA', 'USA', 'USA', 'USA', 'USA', 'USA', 'USA', 'USA', 'USA']
}

df = pd.DataFrame(data)

# Tama√±o de burbuja escalado
df['Bubble_Size'] = (df['Margen_EBITDA'].abs() + 1) * 10

# Colores por mercado
colors = df['Mercado'].map({'BMV': 'green', 'USA': 'royalblue'})

plt.figure(figsize=(12,8))
plt.scatter(df['P_U'], df['Margen_Neto'], s=df['Bubble_Size'], c=colors, alpha=0.6, edgecolors='k')

# Etiquetas
for i in range(len(df)):
    plt.text(df['P_U'][i], df['Margen_Neto'][i], df['Ticker'][i], fontsize=8, ha='center', va='center')

plt.title('Valoraci√≥n (P/U) vs Rentabilidad (Margen Neto) con Margen EBITDA como Tama√±o')
plt.xlabel('P/U (Precio / Utilidad)')
plt.ylabel('Margen Neto (%)')
plt.axvline(x=50, color='green', linestyle='--', alpha=0.5, label='P/U ‚âà 50 (valor justo)')
plt.axhline(y=10, color='gray', linestyle='--', alpha=0.5, label='Margen neto m√≠nimo aceptable (10%)')
plt.legend(['P/U ‚âà 50', 'Margen Neto = 10%', 'Acciones BMV', 'Acciones USA'], loc='best')
plt.grid(True)
plt.show()
```
