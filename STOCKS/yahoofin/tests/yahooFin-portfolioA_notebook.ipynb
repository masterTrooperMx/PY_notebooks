```python
#pip install yfinance
import pandas  as pd #Data manipulation
import numpy as np #Data manipulation

import matplotlib.pyplot as plt # plot graphics
plt.style.use("seaborn")

import yfinance as yf # Yahoo finance
from datetime import datetime
```

```python
# my portfolio
# "BIMBOA.MX", "HERDEZ.MX", "BOLSAA.MX",
# yf.download(tickers = symbol, start = start, end = end)
symbol = ["GENIUS21.MX", "IVVPESOISHRS.MX", "WALMEX.MX", "AAPL", "AMZN", "FDX","NVDA", "XOM"]
start = "2023-01-02" 
# manyways to say now
end = datetime.now().strftime("%Y-%m-%d")

portfolio_df = yf.download(tickers=symbol, start = start, end = end)
portfolio_df.head()
portfolio_df.info()
```

```python
# get AAPL stock info
#portfolio_close_df = portfolio_df['Close']
#aapl_df = portfolio_close_df['AAPL']
aapl_df = portfolio_df['Close']['AAPL']
```

```python
# rest of stocks
amzn_df   = portfolio_df['Close']['AMZN']
fdx_df    = portfolio_df['Close']['FDX']
nvda_df   = portfolio_df['Close']['NVDA']
xom_df    = portfolio_df['Close']['XOM']
gen21_df  = portfolio_df['Close']['GENIUS21.MX']
ivv_df    = portfolio_df['Close']['IVVPESOISHRS.MX']
walmex_df = portfolio_df['Close']['WALMEX.MX']
```

```python
# defining two data sets, daily and monthly
# https://stackoverflow.com/questions/51604967/return-last-date-and-value-each-month-in-pandas
aapl_df_m = aapl_df.groupby(aapl_df.index.month).tail(1)
aapl_df_m
```

```python
# rest of stocks
amzn_df_m   = amzn_df.groupby(amzn_df.index.month).tail(1)
fdx_df_m    = fdx_df.groupby(fdx_df.index.month).tail(1)
nvda_df_m   = nvda_df.groupby(nvda_df.index.month).tail(1)
xom_df_m    = xom_df.groupby(xom_df.index.month).tail(1)
gen21_df_m  = gen21_df.groupby(gen21_df.index.month).tail(1)
ivv_df_m    = ivv_df.groupby(ivv_df.index.month).tail(1)
walmex_df_m = walmex_df.groupby(walmex_df.index.month).tail(1)
```

```python
# creating US df with close values
us_portfolio_df = pd.DataFrame()
us_portfolio_df['AAPL'] = aapl_df.dropna()
us_portfolio_df['AMZN'] = amzn_df.dropna()
us_portfolio_df['FDX']  = fdx_df.dropna()
us_portfolio_df['NVDA'] = nvda_df.dropna()
us_portfolio_df['XOM']  = xom_df.dropna()

# https://numpy.org/doc/stable/reference/generated/numpy.log.html
#us_portfolio_df["AAPL_Returns"] = np.log(aapl_df.div(aapl_df.shift(1)))
us_portfolio_df["AAPL_Returns"] = aapl_df.pct_change()
#us_portfolio_df["AMZN_Returns"] = np.log(amzn_df.div(amzn_df.shift(1)))
us_portfolio_df["AMZN_Returns"] = amzn_df.pct_change()
#us_portfolio_df["FDX_Returns"] = np.log(fdx_df.div(fdx_df.shift(1)))
us_portfolio_df["FDX_Returns"] = fdx_df.pct_change()
#us_portfolio_df["NVDA_Returns"] = np.log(nvda_df.div(nvda_df.shift(1)))
us_portfolio_df["NVDA_Returns"] = nvda_df.pct_change()
#us_portfolio_df["XOM_Returns"] = np.log(xom_df.div(xom_df.shift(1)))
us_portfolio_df["XOM_Returns"] = xom_df.pct_change()

us_portfolio_df
```

```python
# create US dayly df difference with values
us_portfolio_diff_df = pd.DataFrame()
ticks_us = ['AAPL', 'AMZN', 'FDX', 'NVDA', 'XOM']

for t in ticks_us:
    # transform into differences dt
    tmp_l = []
    j = 0
    for i in us_portfolio_df[t].index:
        #print(t, i, j)
        if j > 0:
            # look for consecutive cells
            #print(us_portfolio_df[t].iloc[j-1] - us_portfolio_df[t].iloc[j])
            tmp_l.append(us_portfolio_df[t].iloc[j] - us_portfolio_df[t].iloc[j-1])
        j = j + 1
    us_portfolio_diff_df = pd.concat([us_portfolio_diff_df, pd.DataFrame(tmp_l, columns=[t])], axis=1)

#tmp_portfolio_diff_df = pd.DataFrame(tmp_l, columns=[t])
#us_portfolio_diff_df = pd.concat([us_portfolio_diff_df, tmp_portfolio_diff_df], ignore_index=True, sort=False)

us_portfolio_diff_df
```

```python
# us portfolio monthly info
#aapl_df_m = aapl_df.groupby(aapl_df.index.month).tail(1)
#print(aapl_df_m.mean())
us_portfolio_analysis_df['AAPL']['Month_outcome'] = us_portfolio_m_diff_df['AAPL'].mean()
us_portfolio_analysis_df['AAPL']['Year_outcome']  = us_portfolio_m_diff_df['AAPL'].mean()*12
us_portfolio_analysis_df['AAPL']['Year_var']      = us_portfolio_df["AAPL_Returns"].var()
#us_portfolio_analysis_df['AAPL']['Cov_year_mkt']  = us_portfolio_df["AAPL_Returns"].cov()
us_portfolio_analysis_df['AAPL']['Cov_year_mkt']  = np.sqrt(us_portfolio_df["AAPL_Returns"].var())
#amzn_df_m   = amzn_df.groupby(amzn_df.index.month).tail(1)
us_portfolio_analysis_df['AMZN']['Month_outcome'] = us_portfolio_m_diff_df['AMZN'].mean()
us_portfolio_analysis_df['AMZN']['Year_outcome']  = us_portfolio_m_diff_df['AMZN'].mean()*12
us_portfolio_analysis_df['AMZN']['Year_var']      = us_portfolio_df["AMZN_Returns"].var()
#us_portfolio_analysis_df['AMZN']['Cov_year_mkt']  = us_portfolio_df["AMZN_Returns"].cov()
us_portfolio_analysis_df['AMZN']['Cov_year_mkt']  = np.sqrt(us_portfolio_df["AMZN_Returns"].var())
#fdx_df_m   = fdx_df.groupby(amzn_df.index.month).tail(1)
us_portfolio_analysis_df['FDX']['Month_outcome'] = us_portfolio_m_diff_df['FDX'].mean()
us_portfolio_analysis_df['FDX']['Year_outcome']  = us_portfolio_m_diff_df['FDX'].mean()*12
us_portfolio_analysis_df['FDX']['Year_var']      = us_portfolio_df["FDX_Returns"].var()
#us_portfolio_analysis_df['FDX']['Cov_year_mkt']  = us_portfolio_df["FDX_Returns"].cov()
us_portfolio_analysis_df['FDX']['Cov_year_mkt']  = np.sqrt(us_portfolio_df["FDX_Returns"].var())
#nvda_df_m   = nvda_df.groupby(nvda_df.index.month).tail(1)
us_portfolio_analysis_df['NVDA']['Month_outcome'] = us_portfolio_m_diff_df['NVDA'].mean()
us_portfolio_analysis_df['NVDA']['Year_outcome']  = us_portfolio_m_diff_df['NVDA'].mean()*12
us_portfolio_analysis_df['NVDA']['Year_var']      = us_portfolio_df["NVDA_Returns"].var()
#us_portfolio_analysis_df['NVDA']['Cov_year_mkt']  = us_portfolio_df["NVDA_Returns"].cov()
us_portfolio_analysis_df['NVDA']['Cov_year_mkt']  = np.sqrt(us_portfolio_df["NVDA_Returns"].var())
#xom_df_m    = xom_df.groupby(xom_df.index.month).tail(1)
us_portfolio_analysis_df['XOM']['Month_outcome']  = us_portfolio_m_diff_df['XOM'].mean()
us_portfolio_analysis_df['XOM']['Year_outcome']   = us_portfolio_m_diff_df['XOM'].mean()*12
us_portfolio_analysis_df['XOM']['Year_var']       = us_portfolio_df["XOM_Returns"].var()
#us_portfolio_analysis_df['XOM']['Cov_year_mkt']   = us_portfolio_df["XOM_Returns"].cov()
us_portfolio_analysis_df['XOM']['Cov_year_mkt']   = np.sqrt(us_portfolio_df["XOM_Returns"].var())

us_portfolio_analysis_df
```
