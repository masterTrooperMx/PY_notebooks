```python
import requests
import json

def obtener_datos_api(url):
    try:
        response = requests.get(url)
        
        # Verificar si la respuesta es exitosa
        if response.status_code == 200:
            datos = response.json()
            print("\nüîπ Datos obtenidos de la API üîπ")
            print(json.dumps(datos, indent=4, ensure_ascii=False))
        else:
            print(f"\n‚ùå Error {response.status_code}: No se pudo obtener la informaci√≥n.")
    
    except requests.exceptions.RequestException as e:
        print(f"\n‚ùå Error en la conexi√≥n: {e}")

if __name__ == "__main__":
    url_api = input("üîó Ingresa la URL de la API REST: ")
    obtener_datos_api(url_api)
```
