```python
ls_qnas_2019 = []
ls_qnas_ok = []
ls_qnas_nok = []
for i in qnas:
    ls_qnas = []
    dt_qnas = dt.loc[i]
    print(i)
    print(dt_qnas['nombre_completo'].isin(fd_plan_limpio['nombre_completo']).value_counts())
    ls_qnas.append(i)
    sr_qna_info = dt_qnas['nombre_completo'].isin(fd_plan_limpio['nombre_completo']).value_counts()
    ls_qnas.append(sr_qna_info.to_list())
    ls_qnas_ok.append(ls_qnas[1][0])
    ls_qnas_nok.append(ls_qnas[1][1])
    ls_qnas_2019.append(ls_qnas)
ls_qnas_2019
```
