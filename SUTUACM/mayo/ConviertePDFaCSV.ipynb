vamos a hacer la carga masiva de las quincenas y las vamos a dejar en DF con el nombre de la quincena

```python
# funciones de conversion
# for file in *; do mv "$file" `echo $file | tr ' ' '_'` ; done
def leeArchPDFaDF(n_arch):
    df = tabula.read_pdf(n_arch, stream=True, pages='all', lattice=True)
    return df

# el caso de un archivo con 1 encabezado
def extraeDFdeLsBody(ls_df):
    df = pd.concat([pd.DataFrame(dfi.values) for dfi in ls_df], ignore_index=True)
    return df

def extraeDFdeLsHead(ls_df):
    df_h = pd.concat([pd.DataFrame(dfi.columns).T for dfi in ls_df], ignore_index=True)
    # quita el encabezado
    df_h = df_h.drop(labels=0, axis=0)
    return df_h

# el caso de 1 archivo con multiples encabezados
def pegaTodosDFs(ls_df):
    df = ls_df[0]
    for i in range(len(ls_df)):
        if( i < len(ls_df)):
            df = pd.concat([df,ls_df[i]], ignore_index=True, axis=0)
    return df

# el ciclo principal
ls_nombres = []
dir = "./descuentos/2022/"
for n_arch in os.listdir(dir):
     if n_arch.endswith('.pdf'):
         if(n_arch.startswith('Relación')):
            # leemos el archivo que trae las hojas
            ls_df_tmp = leeArchPDFaDF(dir+n_arch)
            # generamos el DF de quincena
            ls_nombre = n_arch.split('_')
            ls_qna = ls_nombre[8].split('.')
            qna = 'QNA'+ls_qna[0]
            if(qna == 'QNA20190101'and qna == 'QNA20190201'):
                # cuerpo
                df_dina1 = extraeDFdeLsBody(ls_df_tmp)
                # encabezados
                df_dina2 = extraeDFdeLsHead(ls_df_tmp)
                # pega todo
                df_dina = f"{qna}=pd.concat([df_dina1,df_dina2], ignore_index=True, axis=0)"
                exec(df_dina)
                # cambiamos los nombres de las columnas
                df_dina = qna+".rename(columns={0: 'qna', 1: 'nombre', 2: 'puesto', 3: 'descuento'}, inplace=True)"
                exec(df_dina)
            else:
                # pegar todo
                df_dina = f"{qna}=pegaTodosDFs(ls_df_tmp)"
                exec(df_dina)
                # cambiamos los nombres de las columnas
                #df_dina = qna+".rename(columns={0: 'qna', 1: 'nombre', 2: 'puesto', 3: 'descuento'}, inplace=True)"
                #exec(df_dina)
            print(qna)
            ls_nombres.append(qna)
ls_nombres
```

```python
# muestra los DFs creados
for qna in ls_nombres:
    print(qna)
    qna_dina = f"{qna}.info()"
    exec(qna_dina)
    qna_dina = qna+".to_csv('"+dir+qna+".csv')"
    exec(qna_dina)
```
