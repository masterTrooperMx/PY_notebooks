```python
# vamos a pegar las imagenes utilizando la biblioteca ImageMagick
# en un ciclo vamos a reducir todos los QRs
import os, fnmatch

lista_generados = []
dir = "./boletos"
for qr in fnmatch.filter(os.listdir(dir), 'dic_boleto*'):
    arch = qr.split('_')
    try:
        ext = arch[3]
        #print(arch)
        num = ext.split('.')
        lista_generados.append(int(num[0]))
        #print(num[0])
    except ValueError:
        print("No hay numero")
lista_generados.sort()
lista_generados
#len(lista_generados)
```

```python
import pandas as pd

df = pd.read_csv('lista_boletos.csv', index_col='marca_temporal', na_filter= False)
#df.info()
df_boletos_SI = df[df['procede'] == 'SI'][['boleto']]
df_boletos_NO = df[df['procede'] == 'NO'][['boleto']]

lista_SI = df_boletos_SI.values.tolist()
lista_procesados = []
for b in lista_SI:
    lista_procesados.append(b[0])
    print(b[0])
lista_procesados
#len(lista_procesados)
```

```python
# iterar sobre la  la lista de procesados y ver si está en generados
# lo que sobra es lo que falta por generar
lista_FALTA = lista_procesados
for x in lista_FALTA:
    if(x in lista_generados):
        #print(x)
        lista_FALTA.remove(x)
    else:
        print(x)
print(lista_FALTA)
```

```python
def Diff(li1, li2):
    li_dif = [i for i in li1 + li2 if i not in li1 or i not in li2]
    return li_dif
print(Diff(lista_generados, lista_procesados))
```
