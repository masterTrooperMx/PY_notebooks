```python
# import modules
import qrcode
from PIL import Image

# taking image which user wants
# in the QR code center
Logo_link = './sutuacm_redondo__400x400.jpg'

logo = Image.open(Logo_link)

# taking base width
basewidth = 100

# adjust image size
wpercent = (basewidth/float(logo.size[0]))
hsize = int((float(logo.size[1])*float(wpercent)))
logo = logo.resize((basewidth, hsize), Image.ANTIALIAS)
QRcode = qrcode.QRCode(
	error_correction=qrcode.constants.ERROR_CORRECT_H
)

# taking url or text
url = 'https://www.geeksforgeeks.org/'

# adding URL or text to QRcode
QRcode.add_data(url)

# generating QR code
QRcode.make()

# taking color name from user
QRcolor = 'Green'

# adding color to QR code
QRimg = QRcode.make_image(
	fill_color=QRcolor, back_color="white").convert('RGB')

# set size of QR code
pos = ((QRimg.size[0] - logo.size[0]) // 2,
	(QRimg.size[1] - logo.size[1]) // 2)
QRimg.paste(logo, pos)

# save the QR code generated
QRimg.save('gfg_QR.png')

print('QR code generated!')
```

```python
# ponle info y los atributos como esperamos
# import modules
import qrcode
from PIL import Image

data = ['HORACIO HUGO', 'GARCÍA', 'ESPÍNDOLA', '2004030050', '107']

def generaTicket(lista):
    # taking image which user wants
    # in the QR code center
    Logo_link = './sutuacm_redondo__400x400.jpg'

    logo = Image.open(Logo_link)

    # taking base width
    basewidth = 100

    # adjust image size
    wpercent = (basewidth/float(logo.size[0]))
    hsize = int((float(logo.size[1])*float(wpercent)))
    logo = logo.resize((basewidth, hsize), Image.ANTIALIAS)
    QRcode = qrcode.QRCode(
        error_correction=qrcode.constants.ERROR_CORRECT_H
    )

    # taking data and put it right
    data = f"Nombre:{lista[0]} {lista[1]} {lista[2]}, Empleado: {lista[3]}, Boleto: {lista[4]}"

    # adding URL or text to QRcode
    QRcode.add_data(data)

    # generating QR code
    QRcode.make()

    # taking color name from user
    QRcolor = 'orangered'

    # adding color to QR code
    QRimg = QRcode.make_image(
        fill_color=QRcolor, back_color="white").convert('RGB')

    # set size of QR code
    pos = ((QRimg.size[0] - logo.size[0]) // 2,
        (QRimg.size[1] - logo.size[1]) // 2)
    QRimg.paste(logo, pos)

    # save the QR code generated
    QRimg.save(f"./boletos/boleto_{lista[3]}_{lista[4]}.png")

    print('boleto', lista[4])

generaTicket(data)
```

```python
# leer la lista de boletos de un csv
# luego imprimir los que tengan el campo Procede <> de NO
import pandas as pd

df = pd.read_csv('lista_boletos.csv', index_col='marca_temporal', na_filter= False)
df.info()
```

```python
df[['nombre(s)', 'apellido_paterno', 'apellido_materno', 'número_de_empleado','boleto']]
```

```python
# despliega solo los que no deicen NO
df_boletos = df[df['procede'] != 'NO'][['nombre(s)', 'apellido_paterno', 'apellido_materno', 'número_de_empleado','boleto']]
df_boletos
```

```python
# ahora hay que generar los QRs
ls_boletos = df_boletos.values.tolist()
for registro in ls_boletos:
    nombre = f"{registro[0]} {registro[1]} {registro[2]}"
    no_empleado = registro[3]
    boleto = registro[4]
    generaTicket(registro)
    print(nombre, no_empleado, boleto)
```

```python
# vamos a pegar las imagenes utilizando la biblioteca ImageMagick
# en un ciclo vamos a reducir todos los QRs
import os
dir = "./boletos"
for qr in os.listdir(dir):
    # vamos a reducir todos los qrs
    resize_qr = f"convert {dir}/{qr} -resize 478x478 {dir}/{qr}"
    res = os.system(resize_qr)
    print(res)
    # vamos a crear el fondo de los boletos
    create_bkgrd = f"convert -size 1736x478 xc:white {dir}/dic_{qr}"
    res = os.system(create_bkgrd)
    print(res)
    # pegamos la primera parte al boleto
    glue_first = f"composite -geometry  +0+0 ./b_der.png {dir}/dic_{qr} {dir}/dic_{qr}"
    res = os.system(glue_first)
    print(res)
    # pegamos la segunda parte
    glue_second = f"composite -geometry  +868+0 ./b_izq.png {dir}/dic_{qr} {dir}/dic_{qr}"
    res = os.system(glue_second)
    print(res)
    # pegamos el qr
    glue_third = f"composite -geometry  +1086+0 {dir}/{qr} {dir}/dic_{qr} {dir}/dic_{qr}"
    res = os.system(glue_third)
    print(res)
    print(qr)
```
